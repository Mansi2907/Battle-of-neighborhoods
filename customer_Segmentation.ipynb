```python
age_18_25 = df.Age[(df.Age>=18) & (df.Age<=25)]
age_26_35 = df.Age[(df.Age>=26) & (df.Age<=35)]
age_36_45 = df.Age[(df.Age>=36) & (df.Age<=45)]
age_46_55 = df.Age[(df.Age>=46) & (df.Age<=55)]
age_55up = df.Age[df.Age>=56]

agex = ['18-25','26-35','36-45','46-55','55+']
agey = [len(age_18_25.values),len(age_26_35.values),len(age_36_45.values),len(age_46_55.values),len(age_55up.values) ]
plt.figure(figsize = (15,6))
sns.barplot(x=agex, y= agey, palette='mako')
plt.title('Number of Customer and Ages')
plt.xlabel('Age')
plt.ylabel('Number of Customer')
plt.show()
```

```python
score_1_20 = df['Spending Score (1-100)'][(df['Spending Score (1-100)'] >= 1) & (df['Spending Score (1-100)']<= 20)]
score_21_40 = df['Spending Score (1-100)'][(df['Spending Score (1-100)'] >= 21) & (df['Spending Score (1-100)']<= 40)]
score_41_60 = df['Spending Score (1-100)'][(df['Spending Score (1-100)'] >= 41) & (df['Spending Score (1-100)']<= 60)]
score_61_80 = df['Spending Score (1-100)'][(df['Spending Score (1-100)'] >= 61) & (df['Spending Score (1-100)']<= 80)]
score_81_100 = df['Spending Score (1-100)'][(df['Spending Score (1-100)'] >= 81) & (df['Spending Score (1-100)']<= 100)]

scorex = ['1-20','21-40','41-60','61-80','81-100']
scorey = [len(score_1_20.values),len(score_21_40.values),len(score_41_60.values),len(score_61_80.values),len(score_81_100.values)]

plt.figure(figsize = (15,6))
sns.barplot(x=scorex, y= scorey, palette='rocket')
plt.title('Spending score')
plt.xlabel('Score')
plt.ylabel('Number of Customers Having the score')
plt.show()
```

```python
income_30 = df['Annual Income (k$)'][(df['Annual Income (k$)'] >= 0) & (df['Annual Income (k$)']<= 30)]
income_60 = df['Annual Income (k$)'][(df['Annual Income (k$)'] >= 31) & (df['Annual Income (k$)']<= 60)]
income_90 = df['Annual Income (k$)'][(df['Annual Income (k$)'] >= 61) & (df['Annual Income (k$)']<= 90)]
income_120 = df['Annual Income (k$)'][(df['Annual Income (k$)'] >=91) & (df['Annual Income (k$)']<= 120)]
income_150 = df['Annual Income (k$)'][(df['Annual Income (k$)'] >= 121) & (df['Annual Income (k$)']<= 150)]
incomex = ['0k-30k','31k-60k','61k-90k','91k-120k','121k-150k']
incomey = [len(income_30.values),len(income_60.values),len(income_90.values),len(income_120.values),len(income_150.values)]
plt.figure(figsize = (15,6))
sns.barplot(x=incomex, y= incomey, palette='Spectral')
plt.title('Anuual Income')
plt.xlabel('Income')
plt.ylabel('Number of Customers')
plt.show()
```

```python
X1= df.loc[:, ['Age','Spending Score (1-100)']].values

from sklearn.cluster import KMeans
wcss = []
for k in range(1,11):
  kmeans = KMeans(n_clusters=k, init ='k-means++')
  kmeans.fit(X1)
  wcss.append(kmeans.inertia_)
plt.figure(figsize = (12,6))
plt.grid()
plt.plot(range(1,11),wcss, linewidth = 2, color = 'red', marker = '8')
plt.xlabel('K Value')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 4)
label = kmeans.fit_predict(X1)
print(label)
```

```python
print(kmeans.cluster_centers_)
```

```python
plt.scatter(X1[:,0],X1[:,1], c = kmeans.labels_, cmap ='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],color = 'black')
plt.title('Cluster of customers')
plt.xlabel('Age')
plt.ylabel('Spending Score(1-100)')
plt.show()
```
